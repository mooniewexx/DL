1.1.4 Написать функцию, которая для целых значений `n` и `m` будет возвращать тензор размера `n`x`m`, заполненный текстурой размера 2x2, состоящей из следующих значений:

```
0 1

2 3

```

Пример для n = 4 и m = 5:

```
0 1 0 1 0

2 3 2 3 2

0 1 0 1 0

2 3 2 3 2
```

````python
def create_tensor(n, m):
    texture = torch.tensor([[0, 1], [2, 3]])

    tiled_tensor = texture.repeat(n // 2 + 1, m // 2 + 1)[:n, :m]

    return tiled_tensor



tensor_n_m = create_tensor(4, 5)
tensor_n_m
````

1.4.7 Решить матричное уравнение `A*X*B=-C` - найти матрицу X. Где `A = [[-1, 2, 4], [-3, 1, 2], [-3, 0, 1]]`, `B=[[3, -1], [2, 1]]`, `C=[[7, 21], [11, 8], [8, 4]]`.

````python
A = torch.tensor([[-1, 2, 4], [-3, 1, 2], [-3, 0, 1]], dtype=torch.float32)
B = torch.tensor([[3, -1], [2, 1]], dtype=torch.float32)
C = torch.tensor([[7, 21], [11, 8], [8, 4]], dtype=torch.float32)

A_inv = torch.inverse(A)
B_inv = torch.inverse(B)

X = A_inv @ (-C) @ B_inv
X
````
